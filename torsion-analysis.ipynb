## Lowest energy angle

```python
df2 = df.loc[df.name == 'biphenyl-twist']

methods = list(df2.columns[4:])
td = []
for method in methods:
    t0 = df2.theta.loc[df2[method] == df2[method].min()].iloc[0]
    t0_e = df2[method].loc[df2.theta == t0].item()
    d = {}
    d.update({'molecule': df2.name.iloc[0]})
    d.update({'method': method})
    d.update({'theta': t0})
    d.update({'energy': t0_e})
    td.append(d)
        
t_data = pd.DataFrame(td)
```

## Height of barrier

```python
# biphenyl highest and lowest energy angles
b_t1 = -45
b_t2 = 0

# sucrose highest and lowest energy angles
s_t1 = 0
s_t2 = -60
```

```python
df3 = df.loc[df.name == 'biphenyl-twist']

methods = list(df3.columns[3:])
td = []
for method in methods:
    # height of barrier
    hb = abs(df3[method].loc[df3.theta == b_t1].item() - 
             df3[method].loc[df3.theta == b_t2].item())

    d = {}
    d.update({'molecule': df3.name.iloc[0]})
    d.update({'method': method})
    d.update({'height of barrier': hb})
    td.append(d)
        
hb_data = pd.DataFrame(td)
```

```python
hb_data
```

```python
df3 = df.loc[df.name == 'sucrose']

methods = list(df3.columns[4:])
td = []
for method in methods:
    # height of barrier
    hb = abs(df3[method].loc[df3.theta == s_t1].item() - 
             df3[method].loc[df3.theta == s_t2].item())

    d = {}
    d.update({'molecule': df3.name.iloc[0]})
    d.update({'method': method})
    d.update({'height of barrier': hb})
    td.append(d)

pd.options.display.float_format = '{:.4f}'.format
hb_data2 = pd.DataFrame(td)
```

## MAE 2D Torsion Scans

```python
from sklearn.metrics import mean_absolute_error
```

```python
molecules = ['ala-ala', 'aspartame', 'gly-gly']
methods = ['ani1x_energy', 'ani2_energy', 'cnn_energy', 'fchl_energy']

for molecule in molecules:
    d = df.loc[df['name'] == molecule]
    dft = d['energy']
    for method in methods:
        if molecule == 'aspartame':
            if method == 'fchl_energy':
                print(molecule, method, 'FIX')
            else:
                mae = mean_absolute_error(dft, d[method])
                print(molecule, method, '%0.2f' % mae)
        else:
                mae = mean_absolute_error(dft, d[method])
                print(molecule, method, '%0.2f' % mae)
```

```python
molecules = ['ala-ala', 'aspartame', 'gly-gly']
meths = ['energy', 'ani1x_energy', 'ani2_energy', 'cnn_energy', 'fchl_energy']
methods = ['ani1x_energy', 'ani2_energy', 'cnn_energy', 'fchl_energy']

for molecule in molecules:
    d = df.loc[df['name'] == molecule]
    for k in range(len(meths)):
        d['{}-delta'.format(meths[k])] = d[meths[k]] - d[meths[k]].min()
    dft = d['energy-delta']
    for method in methods:
        if molecule == 'aspartame':
            if method == 'fchl_energy':
                print(molecule, method, 'FIX')
            else:
                mae = mean_absolute_error(dft, d['{}-delta'.format(method)])
                print(molecule, method, '%0.2f' % mae)
        else:
                mae = mean_absolute_error(dft, d['{}-delta'.format(method)])
                print(molecule, method, '%0.2f' % mae)
```
