## BAND NN

```python
import sys
sys.path.insert(0, '../scripts/')

from predictor import *
from cpredictor import *
from chemreps.utils.molecule import Molecule
import re
import glob
```

```python
def b_conn_l(mol):
    new_conn = []
    for i in range(mol.n_atom):
        a_conn = []
        for j in range(len(mol.connect)):
            if i+1 in mol.connect[j]:
                if mol.connect[j][0] == i+1:
                    a_conn.append(int(mol.connect[j][1])-1)
                elif mol.connect[j][1] == i+1:
                    a_conn.append(int(mol.connect[j][0])-1)
        new_conn.append(a_conn)
    return new_conn
```

```python
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts
```

### Bond Stertch

```python
molfiles = ['H2', 'HF', 'CO', 'H2O', 'HCCH', 'H2CCH2', 'HCN', 'NH3', 'N2', 'CH4', 'CH3OH', 'C6H6-C', 'C6H6-H', 
            'ala-ala', 'aspartame', 'biphenyl', 'gly-gly', 'sucrose']
modtyp = ['b', 'b', 'b', 'ban', 'band', 'band', 'ban', 'ban', 'b', 'ban', 'band', 'band', 'band',
          'band', 'band', 'band', 'band']
```

```python
results = []
for i in range(len(molfiles)):
    for file in sorted(glob.iglob('../../molecules/stretch/{}/sdf/*.sdf'.format(molfiles[i])), key=numericalSort):
        try:
            pt = file.split('sdf/')[1].split('.')[0]
            mol = Molecule(file)
            species = mol.sym
            coordinates = mol.xyz.tolist()
            bond_connectivity_list = b_conn_l(mol)

            if modtyp[i] == 'b':
                prediction_model = b_prediction_model()
                energy = b_predict_energy(prediction_model,coordinates,species,bond_connectivity_list)
                print(molfiles[i], pt, energy, 'B')

            elif modtyp[i] == 'ban':
                prediction_model = ban_prediction_model()
                energy = ban_predict_energy(prediction_model,coordinates,species,bond_connectivity_list)
                print(molfiles[i], pt, energy, 'BAN')

            elif modtyp[i] == 'band':
                prediction_model = get_default_prediction_model()
                energy = predict_energy(prediction_model,coordinates,species,bond_connectivity_list)
                print(molfiles[i], pt, energy, 'BAND')
            
            d = {}
            d.update({'name': molfiles[i]})
            d.update({'point': pt})
            d.update({'BAND_energy': energy})
            results.append(d)
        
        except:
            print(molfiles[i], pt, 'DID NOT RUN')
            d = {}
            d.update({'name': molfiles[i]})
            d.update({'point': pt})
            d.update({'BAND_energy': np.nan})
            results.append(d)
```

```python
import pandas as pd
df = pd.DataFrame(results, columns=['name', 'point', 'BAND_energy'])
df
```

```python
df.to_csv('../../data/BAND-results.csv', index=False)
```

### Dihedral

```python
molfiles = ['ala-ala', 'aspartame', 'gly-gly', 'sucrose', 'biphenyl-twist']
```

```python
results = []
for i in range(len(molfiles)):
    for file in sorted(glob.iglob('../../molecules/dihedral/{}/sdf/*.sdf'.format(molfiles[i])), key=numericalSort):
        molecule = file.split('.sdf')[0].split('sdf/')[1].split('_')
        name = molecule[0]
        if name == 'sucrose' or name == 'biphenyl-twist':
            phi = 'nan'
            psi = 'nan'
            theta = molecule[1]
        else:
            phi = molecule[1]
            psi = molecule[2]
            theta = 'nan'
        
        try:
            mol = Molecule(file)
            species = mol.sym
            coordinates = mol.xyz.tolist()
            bond_connectivity_list = b_conn_l(mol)

            prediction_model = get_default_prediction_model()
            energy = predict_energy(prediction_model,coordinates,species,bond_connectivity_list)
            print(name, phi, psi, theta, energy, 'BAND')
            
            d = {}
            d.update({'name': name})
            d.update({'phi': phi})
            d.update({'psi': psi})
            d.update({'theta': theta})
            d.update({'BAND_energy': energy})
            results.append(d)
        
        except:
            print(name, phi, psi, theta, 'DID NOT RUN')
            d = {}
            d.update({'name': name})
            d.update({'phi': phi})
            d.update({'psi': psi})
            d.update({'theta': theta})
            d.update({'BAND_energy': np.nan})
            results.append(d)
```

```python
df = pd.DataFrame(results, columns=['name', 'phi', 'psi', 'theta', 'BAND_energy'])
df
```

```python
df.to_csv('../../data/BAND-dihedral-results.csv', index=False)
```
