```python
import pandas as pd
import plotly.express as px
```

```python
ani2 = pd.read_csv('data/ani2-results.csv')
band = pd.read_csv('data/BAND-results.csv')
```

```python
df = ani2.merge(band, on=['name', 'point'])
df
```

```python
df.to_csv('data/data-all.csv', index=False)
```

```python
fig = px.scatter(df, 
                 x='length', 
                 y='ani2_energy',
                 hover_name='name',
                 hover_data=['name', 'point', 'bond', 'length', 'ani2_energy', 'energy', 'BAND_energy'],
                 color='name',
                 template="plotly_white")

fig.show()
fig.write_html('data/ani2.html')
```

```python
fig = px.scatter(df, 
                 x='length', 
                 y='BAND_energy',
                 hover_name='name',
                 hover_data=['name', 'point', 'bond', 'length', 'ani2_energy', 'energy', 'BAND_energy'],
                 color='name',
                 template="plotly_white")

fig.show()
fig.write_html('data/BAND.html')
```

```python
fig = px.scatter(df, 
                 x='length', 
                 y='energy',
                 hover_name='name',
                 hover_data=['name', 'point', 'bond', 'length', 'ani2_energy', 'energy', 'BAND_energy'],
                 color='name',
                 template="plotly_white")

fig.show()
fig.write_html('data/actual.html')
```
