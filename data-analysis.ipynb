```python
import pandas as pd
```

```python
dft = pd.read_csv('data/dft-data.csv')
ani2 = pd.read_csv('data/ani2-results.csv')
ani1x = pd.read_csv('data/ani1x-results.csv')
cnn = pd.read_csv('data/cnn-results.csv')
band = pd.read_csv('data/BAND-results.csv')
fchl = pd.read_csv('data/fchl-results.csv')
bobrfr = pd.read_csv('data/bob-rfr.csv')
bobkrr = pd.read_csv('data/bob-krr.csv')
bobbrr = pd.read_csv('data/bob-brr.csv')
ecfp = pd.read_csv('data/ecfp-rfr.csv')
mmff = pd.read_csv('data/mmff-results.csv')
gaff = pd.read_csv('data/gaff-results.csv')
```

```python
df = pd.merge(dft, ani2, on=['name', 'point'])
df = pd.merge(df, ani1x, on=['name', 'point'])
df = pd.merge(df, cnn, on=['name', 'point'])
df = pd.merge(df, band, on=['name', 'point'])
df = df.merge(bobrfr, on=['name', 'point'])
df = pd.merge(df, bobkrr, on=['name', 'point'])
df = df.merge(bobbrr, on=['name', 'point'])
df = df.merge(fchl, on=['name', 'point'])
df = df.merge(ecfp, on=['name', 'point'])
df = df.merge(mmff, on=['name', 'point'])
df = df.merge(gaff, on=['name', 'point'])
df
```

### Convert energies to kcal/mol

```python
df['energy'] = df['energy'].multiply(23) # eV to kcal/mol
df['ani2_energy'] = df['ani2_energy'].multiply(23) # eV to kcal/mol
df['ani1x_energy'] = df['ani1x_energy'].multiply(627.50947) # hartree to kcal/mol
df['fchl_energy'] = df['fchl_energy'].multiply(627.50947) # hartree to kcal/mol
df['bobrfr_energy'] = df['bobrfr_energy'].multiply(627.50947) # hartree to kcal/mol
df['bobkrr_energy'] = df['bobkrr_energy'].multiply(627.50947) # hartree to kcal/mol
df['bobbrr_energy'] = df['bobbrr_energy'].multiply(627.50947) # hartree to kcal/mol
df['ecfprfr_energy'] = df['ecfprfr_energy'].multiply(627.50947) # hartree to kcal/mol

df
```

```python
df.to_csv('data/data-all.csv', index=False)
```
