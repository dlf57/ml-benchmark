Find r<sub>0 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/data-all.csv')
df = df.fillna(np.nan)
df
```

```python
names = list(np.unique(df['name']))
names.remove('HF') # drop HF for now
print(names)
```

```python
r0 = []
for name in names:
    ndf = df[df['name'] == name]
    min_e = min(ndf['energy'])
    min_a = min(ndf['ani2_energy'])
    min_a1x = min(ndf['ani1x_energy'])
    if name == 'CO':
        # CO goes down after 2.5 angstroms 
        min_a = min(ndf['ani2_energy'][:15])
        min_a1x = min(ndf['ani1x_energy'][:15])
    if name == 'H2':
        min_a1x = min(ndf['ani1x_energy'][:15])
    if name == 'CH4':
        min_a1x = min(ndf['ani1x_energy'][:15])
    if name == 'H2O':
        min_a1x = min(ndf['ani1x_energy'][:15])
    if name == 'NH3':
        min_a1x = min(ndf['ani1x_energy'][:15])
        
    min_gaff = min(ndf['gaff_energy'][:10])
    min_cnn = min(ndf['cnn_energy'])
    min_fchl = min(ndf['fchl_energy'])
    min_bob = min(ndf['bobrfr_energy'])
    min_bobk = min(ndf['bobkrr_energy'])
    min_bobb = min(ndf['bobbrr_energy'])
    min_ecfp = min(ndf['ecfprfr_energy'])
    min_mmff = min(ndf['mmff_energy'][:9])
    if name == 'sucrose':
        min_b == 'nan'
    else:
        min_b = min(ndf['BAND_energy'].dropna())
    
    pt_e = ndf.loc[df['energy'] == min_e, 'point'].iloc[0]
    pt_a = ndf.loc[df['ani2_energy'] == min_a, 'point'].iloc[0]
    pt_mmff = ndf.loc[df['mmff_energy'] == min_mmff, 'point'].iloc[0]
    pt_gaff = ndf.loc[df['gaff_energy'] == min_gaff, 'point'].iloc[0]
    
    pt_a1x = ndf.loc[df['ani1x_energy'] == min_a1x, 'point'].iloc[0]
    pt_cnn = ndf.loc[df['cnn_energy'] == min_cnn, 'point'].iloc[0]
    
    pt_fchl = ndf.loc[df['fchl_energy'] == min_fchl, 'point'].iloc[0]
    pt_bob = ndf.loc[df['bobrfr_energy'] == min_bob, 'point'].iloc[0]
    pt_bobk = ndf.loc[df['bobkrr_energy'] == min_bobk, 'point'].iloc[0]
    pt_bobb = ndf.loc[df['bobbrr_energy'] == min_bobb, 'point'].iloc[0]
    pt_ecfp = ndf.loc[df['ecfprfr_energy'] == min_ecfp, 'point'].iloc[0]
    if name == 'sucrose':
        pt_b == 'nan'
    else:
        pt_b = ndf.loc[df['BAND_energy'] == min_b, 'point'].iloc[0]

    l_e = ndf.loc[df['energy'] == min_e, 'length'].iloc[0]
    l_a = ndf.loc[df['ani2_energy'] == min_a, 'length'].iloc[0]
    l_mmff = ndf.loc[df['mmff_energy'] == min_mmff, 'length'].iloc[0]
    l_gaff = ndf.loc[df['gaff_energy'] == min_gaff, 'length'].iloc[0]
    
    l_a1x = ndf.loc[df['ani1x_energy'] == min_a1x, 'length'].iloc[0]
    l_cnn = ndf.loc[df['cnn_energy'] == min_cnn, 'length'].iloc[0]
    
    l_fchl = ndf.loc[df['fchl_energy'] == min_fchl, 'length'].iloc[0]
    l_bob = ndf.loc[df['bobrfr_energy'] == min_bob, 'length'].iloc[0]
    l_bobk = ndf.loc[df['bobkrr_energy'] == min_bobk, 'length'].iloc[0]
    l_bobb = ndf.loc[df['bobbrr_energy'] == min_bobb, 'length'].iloc[0]
    l_ecfp = ndf.loc[df['ecfprfr_energy'] == min_ecfp, 'length'].iloc[0]
    if name == 'sucrose':
        l_b == 'nan'
    else:
        l_b = ndf.loc[df['BAND_energy'] == min_b, 'length'].iloc[0]
    
    d = {}
    d.update({'molecule': name})
    d.update({'point-dft': pt_e})
    d.update({'length-dft': l_e})
    d.update({'minE-dft': min_e})
    d.update({'point-ani2': pt_a})
    d.update({'length-ani2': l_a})
    d.update({'minE-ani2': min_a})
    d.update({'point-ani1x': pt_a1x})
    d.update({'length-ani1x': l_a1x})
    d.update({'minE-ani1x': min_a1x})
    d.update({'point-cnn': pt_cnn})
    d.update({'length-cnn': l_cnn})
    d.update({'minE-cnn': min_cnn})
    d.update({'point-band': pt_b})
    d.update({'length-band': l_b})
    d.update({'minE-band': min_b})
    d.update({'point-fchl': pt_fchl})
    d.update({'length-fchl': l_fchl})
    d.update({'minE-fchl': min_fchl})
    d.update({'point-bobrfr': pt_bob})
    d.update({'length-bobrfr': l_bob})
    d.update({'minE-bobrfr': min_bob})
    d.update({'point-bobkrr': pt_bobk})
    d.update({'length-bobkrr': l_bobk})
    d.update({'minE-bobkrr': min_bobk})
    d.update({'point-bobbrr': pt_bobb})
    d.update({'length-bobbrr': l_bobb})
    d.update({'minE-bobbrr': min_bobb})
    d.update({'point-ecfprfr': pt_ecfp})
    d.update({'length-ecfprfr': l_ecfp})
    d.update({'minE-ecfprfr': min_ecfp})
    d.update({'point-mmff': pt_mmff})
    d.update({'length-mmff': l_mmff})
    d.update({'minE-mmff': min_mmff})
    d.update({'point-gaff': pt_gaff})
    d.update({'length-gaff': l_gaff})
    d.update({'minE-gaff': min_gaff})
    r0.append(d)
```

```python
rdf = pd.DataFrame(r0, columns=['molecule', 
                                'point-dft', 'length-dft', 'minE-dft', 
                                'point-ani2', 'length-ani2', 'minE-ani2',
                                'point-ani1x', 'length-ani1x', 'minE-ani1x',
                                'point-cnn', 'length-cnn', 'minE-cnn',
                                'point-band', 'length-band', 'minE-band',
                                'point-fchl', 'length-fchl', 'minE-fchl',
                                'point-bobrfr', 'length-bobrfr', 'minE-bobrfr',
                                'point-bobkrr', 'length-bobkrr', 'minE-bobkrr',
                                'point-bobbrr', 'length-bobbrr', 'minE-bobbrr',
                                'point-ecfprfr', 'length-ecfprfr', 'minE-ecfprfr',
                                'point-mmff', 'length-mmff', 'minE-mmff',
                                'point-gaff', 'length-gaff', 'minE-gaff'])
rdf
```

```python
rdf.to_csv('data/r0.csv', index=False)
```

### MAE at r<sub>0

```python
from sklearn.metrics import mean_absolute_error
```

```python
# mean_absolute_error(y_true, y_pred)
res = []
for name in names:
    d = {}
    for method in list(df.columns)[5:]:
        dftE = df[df.name == name].energy[df.point == 5].item()
        methE = df[df.name == name][method][df.point == 5].item()
        try:
            mae = mean_absolute_error([dftE], [methE])
            d.update({'molecule': name})
            d.update({method: mae})
        except:
            d.update({'molecule': name})
            d.update({method: np.nan})
    res.append(d)
```

```python
dmae = pd.DataFrame(res)
dmae
```
