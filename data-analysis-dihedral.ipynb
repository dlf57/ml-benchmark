```python
import pandas as pd
```

```python
dft = pd.read_csv('data/dft-data-dihedral.csv')
ani2 = pd.read_csv('data/ani2-dihedral-results.csv')
band = pd.read_csv('data/BAND-dihedral-results.csv')
# fchl = pd.read_csv('data/fchl-dihedral-results.csv')
bobrfr = pd.read_csv('data/bob-rfr-dihedral.csv')
bobkrr = pd.read_csv('data/bob-krr-dihedral.csv')
bobbrr = pd.read_csv('data/bob-brr-dihedral.csv')
mmff = pd.read_csv('data/mmff-dihedral-results.csv')
gaff = pd.read_csv('data/gaff-dihedral-results.csv')
```

```python
df = pd.merge(dft, ani2, on=['name', 'phi', 'psi', 'theta'])
df = pd.merge(df, band, on=['name', 'phi', 'psi', 'theta'])
df = df.merge(bobrfr, on=['name', 'phi', 'psi', 'theta'])
df = pd.merge(df, bobkrr, on=['name', 'phi', 'psi', 'theta'])
df = df.merge(bobbrr, on=['name', 'phi', 'psi', 'theta'])
# df = df.merge(fchl, on=['name', 'phi', 'psi', 'theta'])
df = df.merge(mmff, on=['name', 'phi', 'psi', 'theta'])
df = df.merge(gaff, on=['name', 'phi', 'psi', 'theta'])
df
```

### Convert BoB energies to eV

```python
# df['fchl_energy'] = df['fchl_energy'].multiply(27.2114)
df['bobrfr_energy'] = df['bobrfr_energy'].multiply(27.2114)
df['bobkrr_energy'] = df['bobkrr_energy'].multiply(27.2114)
df['bobbrr_energy'] = df['bobbrr_energy'].multiply(27.2114)
df
```

```python
df.to_csv('data/data-all-dihedral.csv', index=False)
```
